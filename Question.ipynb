### Escolha de algoritmos

1) Escreva em tópicos as etapas do algoritmo de K-médias até sua convergência.
- Inicialização aleatória de um ponto próximo aos pontos
- São calculadas as distâncias entre os pontos e o centro, e o cluster é recalculao
- É feita uma iteração N vezes até que todos os pontos tenham a menor distancia possível entre o centro, e todos estejam dentro de N clusters
- Os centroids não são necessáriamente um ponto existente, mas sim, um ponto em que a distancia entre todos os pontos do cluster seja a menor possível

2) O algoritmo de K-médias converge até encontrar os centróides que melhor descrevem os clusters encontrados (até o deslocamento entre as interações dos centróides ser mínimo). Lembrando que o centróide é o baricentro do cluster em questão e não representa, em via de regra, um dado existente na base. Refaça o algoritmo apresentado na questão 1 a fim de garantir que o cluster seja representado pelo dado mais próximo ao seu baricentro em todas as iterações do algoritmo.
Obs: nesse novo algoritmo, o dado escolhido será chamado medóide.
- No caso do Medoid, o ponto central é um ponto real dentro do cluster, desta forma, todos os pontos do cluster tem a menor distância em relação a um ponto central do modelo.
- A diferença entre um modelo e outro é que no K-Medoid um ponto aleatório é escolhido, ao invés de um ponto inexistende do K-Means

3) O algoritmo de K-médias é sensível a outliers nos dados. Explique.
- Outliers são sempre um problema. No caso do K-Means, um outliers vai afastar o centroid do cluster na sua direação, fazendo com que o centro, e o cluster, fiquei muito afastados. Desta forma, existe uma probabilidade de ter uma quantidade de pontos menor dentro do cluster

4) Por que o algoritmo de DBScan é mais robusto à presença de outliers?
- O DB-Scan define os cluster através da densidade de pontos. Dessa forma, os pontos que são outlaeiers são considerados ruídos e esses pontos não sao classificados, são deixados de lado, e não são atribuidos a nenhum cluster.
